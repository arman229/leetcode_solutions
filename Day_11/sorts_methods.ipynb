```python
# 1 Bubble Sort Algorithm
def bubble_sort(array):
    n = len(array)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if array[j]>array[j+1]:
                array[j],array[j+1] = array[j+1],array[j]
                swapped = True
        if not swapped:
                break
    return array 

print(bubble_sort([2, 1, 3, 4, 5]))
# time complexity: O(n^2) in worst case, O(n) in best case
# space complexity: O(1) in-place sorting algorithm
```

```python
# 2 Selection Sort Algorithm
def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

print(selection_sort([9,3,4,5,6]))
# time complexity: O(n^2) in worst case, O(n^2) in best case
# space complexity: O(1) in-place sorting algorithm
```

```python
# 3 Insertion Sort Algorithm
def insertion_sort(arr):
    n = len(arr)
    for i in range(1,n):
        key =  arr[i]
        j = i - 1
        while j>=0 and arr[j] > key:
            arr[j+1] = arr[j] 
            j -=1
        arr[j+1] = key
    return arr
print(insertion_sort([9,3,4,5,6]))        

# time complexity: O(n^2) in worst case|average case, O(n) in best case(sorted array)
# space complexity: O(1) in-place sorting algorithm
```

```python
# 4 Counting Sort Algorithm
def counting_sort(arr):
    if not arr:
        return []
    array_len = max(arr)
    counting = [0]*(array_len + 1)
    
    for i in arr: 
        counting[i] +=1 
             
    final_result = []
    for i in range(len(counting)):
        
        final_result.extend([i]*counting[i])
    return final_result

print(counting_sort([9,3,4,5,6,2,3,4]))

# time complexity: O(n+k) where n is the number of elements in the input array and k is the range of the input values
# space complexity: O(n+k) where n is the number of elements in the input array and k is the range of the input values
```

```python
# 5 Quick Sort Algorithm
def partition(arr,low,high):
    pivot = arr[high]
    i = low - 1
    for j in range(low,high):
        if arr[j]<=pivot:
            i+=1
            arr[i],arr[j] = arr[j],arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i+1

def quick_sort(arr,low=0,high=None):
    if high is None:
        high = len(arr) - 1
    if low < high:
        par = partition(arr,low,high)
        quick_sort(arr,low, par-1)
        quick_sort(arr,par+1,high)
my_array = [64, 66, 25, 12, 22, 11, 90, 5]
quick_sort(my_array)
print("Sorted array:", my_array)
```

```python
# 6 Merge Sort Algorithm

def final_merge(left,right):
    final_resp = []
    i=j=0
    while i<len(left) and j<len(right):
        if left[i] < right[j]:
            final_resp.append(left[i])
            i+=1
        else:
            final_resp.append(right[j])
            j+=1
    final_resp.extend(left[i:])
    final_resp.extend(right[j:])
    return final_resp
def merge_sort(arr):
    n = len(arr)
    if n<=1:
        return arr
    middle = n//2
    left_side = merge_sort(arr[:middle])
    right_side = merge_sort(arr[middle:])
    return final_merge(left_side,right_side)
print(merge_sort([8,4,5,7,8]))
```
