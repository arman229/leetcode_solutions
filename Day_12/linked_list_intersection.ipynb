 
## 160 Intersection of Two Linked Lists  

---

Given the heads of two singly linked lists `headA` and `headB`, return the node at which the two lists intersect. If the two linked lists have no intersection, return `null`.

For example, the following two linked lists begin to intersect at node `c1`:

---

The test cases are generated such that there are no cycles in the entire linked structure.

Note that the linked lists must retain their original structure after the function returns.

---

### Custom Judge Description

The inputs are given as follows (your program does not receive these directly):

- `intersectVal` → value of the intersected node (0 if no intersection)
- `listA` → first linked list
- `listB` → second linked list
- `skipA` → number of nodes skipped in `listA` before intersection
- `skipB` → number of nodes skipped in `listB` before intersection

The judge constructs the linked lists accordingly and passes `headA` and `headB` to your function.

---

### Example 1

**Input:**  
```text
intersectVal = 8
listA = [4,1,8,4,5]
listB = [5,6,1,8,4,5]
skipA = 2
skipB = 3
````

**Output:**

```text id="xg4l2p"
Intersected at '8'
```

**Explanation:**
Both lists intersect at node with value `8`.
Nodes before intersection differ, but the node reference is the same.

### Constraints

* `m` = number of nodes in `listA`
* `n` = number of nodes in `listB`
* `1 <= m, n <= 3 * 10^4`
* `1 <= Node.val <= 10^5`
* `0 <= skipA <= m`
* `0 <= skipB <= n`
* `intersectVal == 0` if no intersection
* If intersection exists:
  `intersectVal == listA[skipA] == listB[skipB]`

---

### Follow-up

Can you solve it in:

* **O(m + n)** time complexity
* **O(1)** extra space?

```


`````python
# Definition for singly-linked list.
from typing import Optional
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        pot_a, pot_b = headA, headB
        while pot_a != pot_b:
            pot_a = pot_a.next if pot_a else headB
            pot_b = pot_b.next if pot_b else headA
        return pot_a
        
`````

### Complexity Analysis 
- Time Complexity: O(n + m), where n and m denote the lengths of list A and B, respectively. This is because pointers linearly traverse both linked lists sequentially. 
- Space Complexity: O(1)
