
## 35. Search Insert Position  


---

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be inserted in order.

You must write an algorithm with **O(log n)** runtime complexity.

---

### Example 1

**Input:**  
```text
nums = [1,3,5,6], target = 5
````

**Output:**

```text id="llr1m0"
2
```

```

### Example 3

**Input:**

```text id="4zjlwm"
nums = [1,3,5,6], target = 7
```

**Output:**

```text id="0axx2d"

```

---

### Constraints

* `1 <= nums.length <= 10^4`
* `-10^4 <= nums[i] <= 10^4`
* `nums` contains distinct values sorted in ascending order
* `-10^4 <= target <= 10^4`

```



`````python
# By Using Brute Force
from typing import List
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        for i in range(len(nums)):
            if nums[i] >= target:
                return i
        return len(nums)

        
        
`````

### Complexity Analysis 
- Time Complexity: O(n)  
- Space Complexity: O(1)  

`````python
# By Using the Binary Search Algorithm
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums)
        while left < right:
            mid = (left + right) //2
            if nums[mid] < target:
                left = mid + 1
            else:
                right = mid
        return left
        
        
`````

### Complexity Analysis 
- Time Complexity: O(log(n))  
- Space Complexity: O(1)
