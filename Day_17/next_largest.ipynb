## Next Largest Number to the Right

Given an integer array `nums`, return an output array `res` where, for each value `nums[i]`, `res[i]` is the first number to the right that is larger than `nums[i]`.

If no larger number exists to the right of `nums[i]`, set `res[i]` to `-1`.

---

### Example

**Input:**  
```text
nums = [5, 2, 4, 6, 1]
````

**Output:**

```text
[6, 4, 6, -1, -1]
```

**Explanation:**

| nums[i] | Next Larger Number |
| ------- | ------------------ |
| 5       | 6                  |
| 2       | 4                  |
| 4       | 6                  |
| 6       | -1                 |
| 1       | -1                 |

Therefore:

```text
res = [6, 4, 6, -1, -1]
```

---

### Constraints

* `1 <= nums.length <= 10^5`
* `-10^9 <= nums[i] <= 10^9`

```
```


`````python
def next_largest(nums):
    stack = []
    res = [0]*len(nums)
    for i in range(len(nums)-1, -1, -1):
        while stack and stack[-1] <= nums[i]:
            stack.pop()
        res[i] = stack[-1] if stack else -1
        stack.append(nums[i])
    return res
print(next_largest([5, 2, 4, 6, 1]))
    
`````

### Complexity Analysis 
- Time Complexity: O(n)  
- Space Complexity: O(n)  

`````python
def next_largest(nums):
    stack = []
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] < nums[j]:
                stack.append(nums[j])
                break
        else:
            stack.append(-1)
    return stack
print(next_largest([5, 2, 4,4,2,1,55, 6, 1] ))
`````
