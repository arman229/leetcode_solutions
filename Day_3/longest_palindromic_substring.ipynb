 
## 5. Longest Palindromic Substring  
 
---

Given a string `s`, return the **longest palindromic substring** in `s`.

---

### Example 1

**Input:**  
```

s = "babad"

```
**Output:**  
```

"bab"

```
**Explanation:**  
"aba" is also a valid answer.

---

### Example 2

**Input:**  
```

s = "cbbd"

```
**Output:**  
```

"bb"

```

---

### Constraints
- `1 <= s.length <= 1000`
- `s` consists of only digits and English letters.
```


````python
class Solution: 
    def longestPalindrome(self, s: str) -> str:
        len_s = len(s)
        if len_s<=1:
            return s
        def extend(left,right):
            while left>=0 and right<len_s and s[left]==s[right]:
                left -=1
                right +=1
            resp= s[left+1:right]
            
            return resp

        final_resp = ""
        for i in range(len_s):
            odd_resp = extend(i,i)
            even_resp = extend(i,i+1)
            print(odd_resp,even_resp)
            
            final_resp_len = len(final_resp)
        
            odd_resp_len = len(odd_resp)
            even_resp_len = len(even_resp)
            if final_resp_len < odd_resp_len:
                final_resp=odd_resp
            if final_resp_len < even_resp_len:
                final_resp=even_resp
        return final_resp



sol =Solution()
test_str = 'abaccabacd'
print(sol.longestPalindrome(test_str))
````
