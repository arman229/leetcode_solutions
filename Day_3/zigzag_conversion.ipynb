## 6. Zigzag Conversion  
---

The string `"PAYPALISHIRING"` is written in a **zigzag pattern** on a given number of rows.  
For example:

```

P   A   H   N
A P L S I I G
Y   I   R

```

Reading line by line â†’ `"PAHNAPLSIIGYIR"`

Write a function that performs this conversion:

```

string convert(string s, int numRows);

```

---

### Example 1

**Input:**  
```

s = "PAYPALISHIRING", numRows = 3

```
**Output:**  
```

"PAHNAPLSIIGYIR"

```

---

### Example 2

**Input:**  
```

s = "PAYPALISHIRING", numRows = 4

```
**Output:**  
```

"PINALSIGYAHRPI"

```

**Explanation:**
```

P     I    N
A   L S  I G
Y A   H R
P     I

```

---

### Example 3

**Input:**  
```

s = "A", numRows = 1

```
**Output:**  
```

"A"

```

---

### Constraints
- `1 <= s.length <= 1000`
- `s` consists of English letters (upper-case and lower-case), `,` and `.`  
- `1 <= numRows <= 1000`
```


````python


from typing import List, Any
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        len_s = len(s)
        if len_s <=1 or numRows >=len_s or numRows ==1:
            return s
        down = True
        cur_location = 0
        static_store:List[Any] = [[] for _ in range(numRows)]
        for i in range(len_s):
            static_store[cur_location].append(s[i])
            if down:
                cur_location +=1
            else:
                cur_location -=1
            if cur_location == 0 or cur_location == numRows - 1  :
                down = not down
        final_resp = ""
        for item in static_store:
            final_resp +="".join(item)
        return final_resp
        
sol = Solution()
print(sol.convert("AB", 1))


````
