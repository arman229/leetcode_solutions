## 8. String to Integer (atoi)  
---

Implement the `myAtoi(string s)` function, which converts a string to a **32-bit signed integer**.

### Algorithm for `myAtoi(string s)`

1. **Whitespace**  
   Ignore any leading whitespace (`" "`).

2. **Signedness**  
   Determine the sign by checking if the next character is `'-'` or `'+'`.  
   Assume the number is positive if neither is present.

3. **Conversion**  
   Read the integer by skipping leading zeros until a non-digit character is encountered or the end of the string is reached.  
   If no digits are read, return `0`.

4. **Rounding**  
   If the integer is outside the 32-bit signed integer range  
   `[-2³¹, 2³¹ - 1]`, clamp it to the range:
   - Less than `-2³¹` → return `-2³¹`
   - Greater than `2³¹ - 1` → return `2³¹ - 1`

5. **Return**  
   Return the final integer result.

---

### Example 1

**Input:**  
```

s = "42"

```

**Output:**  
```

42

```

**Explanation:**
```

Step 1: "42"   (no leading whitespace)
^
Step 2: "42"   (no sign detected)
^
Step 3: "42"   ("42" is read)
^

```

---

### Example 2

```

```

```

```

Step 1: "   -042"   (leading whitespace ignored)
^
Step 2: "   -042"   ('-' detected → negative number)
^
Step 3: "   -042"   ("042" read; leading zero ignored)
^

```

---

### Constraints
- `0 <= s.length <= 200`
- `s` consists of:
  - English letters (lowercase and uppercase)
  - Digits (`0-9`)
  - Space `' '`
  - `'+'`, `'-'`, and `'.'`
```


````python
class Solution:
    def myAtoi(self, s: str) -> int:
        MAX_VAL = 2**31-1
        MIN_VAL = -2**31
        n = len(s)
        i=0
        while i<n and s[i] == ' ':
            i+=1
        sign = 1
        if i<n and s[i] in ["+","-"]:
            if s[i]=='-':
                sign = -1 
            i+=1
        final_num = 0
        limit = MAX_VAL if sign == 1 else -MIN_VAL
        while i<n and s[i].isdigit():
            digit = int(s[i])
            if final_num > limit//10 or (final_num==limit//10 and digit > limit%10) :
                return MAX_VAL if sign==1 else MIN_VAL
                
            i+=1
            final_num = final_num*10 + digit
        return final_num*sign

````
