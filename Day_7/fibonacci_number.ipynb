 
 
## 509. Fibonacci Number   

---

The Fibonacci numbers, commonly denoted as `F(n)`, form a sequence such that each number is the sum of the two preceding ones, starting from `0` and `1`.

Formally:

- `F(0) = 0`
- `F(1) = 1`
- `F(n) = F(n - 1) + F(n - 2)` for `n > 1`

Given an integer `n`, calculate `F(n)`.

---

### Example 1

**Input:**  
```

n = 2

```

**Output:**  
```

1

```

**Explanation:**  
```

F(2) = F(1) + F(0) = 1 + 0 = 1

```

---

### Example 2

**Input:**  
```

n = 3

```

**Output:**  
```

2

```

**Explanation:**  
```

F(3) = F(2) + F(1) = 1 + 1 = 2

```

---

### Example 3

**Input:**  
```

n = 4

```

**Output:**  
```

3

```

**Explanation:**  
```

F(4) = F(3) + F(2) = 2 + 1 = 3

```

---

### Constraints
- `0 <= n <= 30`
```


````python
class Solution:
    def fib(self, n: int) -> int:
        memo = [-1]*(n+1) 
        def cal(i: int):
            if i<=1:
                return i
            if memo[i] != -1:
                return memo[i]
            memo[i] = cal(i-1) + cal(i-2)
            return memo[i]
        return cal(n)
````

````python
class Solution:
    def fib(self, n: int) -> int:
        if n<2:
            return n

        resp = [0]*(n+1)
        resp[0]=0
        resp[1]=1
        for i in range(2,n+1):
            resp[i] = resp[i-1]+resp[i-2]
        return resp[-1]
````
