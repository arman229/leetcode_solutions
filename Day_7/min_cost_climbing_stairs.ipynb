 
## 746. Min Cost Climbing Stairs  
---

You are given an integer array `cost`, where `cost[i]` is the cost of the *i-th* step on a staircase. Once you pay the cost, you can either climb **one step** or **two steps**.

You can start from either:
- Step with index `0`, or
- Step with index `1`.

Return the **minimum cost** to reach the top of the floor.

---

### Example 1

**Input:**  
```

cost = [10, 15, 20]

```

**Output:**  
```

15

```

**Explanation:**  
You start at index `1`:
- Pay `15` and climb two steps to reach the top.

Total cost = `15`.

---

### Example 2

**Input:**  
```

cost = [1,100,1,1,1,100,1,1,100,1]

```

**Output:**  
```

6

```

**Explanation:**  
You start at index `0`:
- Pay `1`, climb two steps to index `2`
- Pay `1`, climb two steps to index `4`
- Pay `1`, climb two steps to index `6`
- Pay `1`, climb one step to index `7`
- Pay `1`, climb two steps to index `9`
- Pay `1`, climb one step to the top

Total cost = `6`.

---

### Constraints
- `2 <= cost.length <= 1000`
- `0 <= cost[i] <= 999`
```


````python
from typing import List
````

````python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        n = len(cost)
        resp = [0]*(n+1)
        for i in range(2,n+1):
            resp[i] = min(resp[i-2]+cost[i-2],resp[i-1]+cost[i-1])
        return resp[n]
````

````python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        n = len(cost)
        cur,prev = 0,0
        for i in range(2,n+1):
            prev,cur =cur, min(prev+cost[i-2],cur+cost[i-1])
        return cur
````

````python
# Memoization
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        n = len(cost)
        resp = {0:0,1:0}
        def cal(i):
            if i in resp:
                return resp[i]
            else:
                resp[i]= min(cost[i-2]+cal(i-2),cost[i-1]+cal(i-1))
                return resp[i]
        return cal(n)
sol = Solution()
print(sol.minCostClimbingStairs([10, 15, 20]))  
````
