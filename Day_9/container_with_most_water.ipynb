## 11. Container With Most Water  

---

You are given an integer array `height` of length `n`.

There are `n` vertical lines drawn such that the two endpoints of the `i-th` line are:

- `(i, 0)`
- `(i, height[i])`

Find two lines that, together with the x-axis, form a container such that the container contains the **most water**.

Return the **maximum amount of water** a container can store.

> Note: You may not slant the container.

---

### Example 1

**Input:**  
```text
height = [1,8,6,2,5,4,8,3,7]
````

**Output:**

```text
49
```

**Explanation:**

The vertical lines are represented by the array:

```text
[1,8,6,2,5,4,8,3,7]
```

The maximum area of water the container can store is `49`.

---

### Example 2

**Input:**

```text
height = [1,1]
```

**Output:**

```text
1
```

---

### Constraints

* `n == height.length`
* `2 <= n <= 10^5`
* `0 <= height[i] <= 10^4`

```

`````python

from typing import List
`````

`````python
# Efficient Solution using Two Pointers
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left, right = 0,len(height)-1
        assume_area = 0
        while left < right:
            find_area = min(height[left], height[right])*(right-left)
            assume_area = max(assume_area,find_area)
            if height[left] < height[right]:
                left+=1
            elif height[left] > height[right]:
                right-=1
            else:
                left+=1
                right-=1
        return assume_area
        


        
`````

### Complexity Analysis 
- Time Complexity: O(N)  
- Space Complexity: O(1)
