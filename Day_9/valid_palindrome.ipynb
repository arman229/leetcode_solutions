## 125. Valid Palindrome  
---

A phrase is a **palindrome** if, after:

1. Converting all uppercase letters into lowercase letters  
2. Removing all non-alphanumeric characters  

it reads the same forward and backward.

Alphanumeric characters include **letters and numbers**.

Given a string `s`, return `true` if it is a palindrome, or `false` otherwise.

---

### Example 1

**Input:**  
```text
s = "A man, a plan, a canal: Panama"
````

**Output:**

```text
true
```

**Explanation:**

```text
"amanaplanacanalpanama" is a palindrome.
```

---

### Example 2

**Input:**

```text
s = "race a car"
```

**Output:**

```text
false
```

**Explanation:**

```text
"raceacar" is not a palindrome.
```

---

### Example 3

**Input:**

```text
s = " "
```

**Output:**

```text
true
```

**Explanation:**
After removing non-alphanumeric characters, `s` becomes an empty string `""`.

Since an empty string reads the same forward and backward, it is a palindrome.

---

### Constraints

* `1 <= s.length <= 2 * 10^5`
* `s` consists only of printable ASCII characters

```


`````python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        lowerstring = s.lower() 
        left,right = 0,len(lowerstring)-1
        while left < right:
            while left<right and not lowerstring[left].isalnum():
                left +=1
            while left<right and not lowerstring[right].isalnum():
                right -=1
            if lowerstring[left] != lowerstring[right]:
                return False
            left +=1
            right -=1

        return True
### Complexity Analysis
# - Time Complexity: O(N)
# - Space Complexity: O(N)
`````
