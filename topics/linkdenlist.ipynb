```python
#Creating Doubly Linked list
from tkinter.messagebox import RETRY


class Node:
    def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None
class DoublyList:
    def __init__(self):
        self.head = None
    def append(self,data):
        create_node = Node(data)
        current = self.head
        if not current:
            self.head = create_node
            return
        
        while current.next:
            current = current.next
        current.next = create_node
        create_node.prev = current 
        
    def traverse(self):
        current = self.head
        if not current:
            print('Doubly Linkden List is Empty')
            return
        while current:
            print('Data is ',current.data,current.prev,current.next)
            current = current.next
            
doubly_list = DoublyList()
print('+++++++++++++++++++++++++')
doubly_list.traverse()
for i in range(4):
    doubly_list.append(i)    
print('+++++++++++++++++++++++++')
doubly_list.traverse()
```

```python
# Circular Singly Linked List

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
class CircularSLinkedList:
    def __init__(self):
        self.head = None
    def append(self,data):
        create_node = Node(data)
        current = self.head
        if not current:
            self.head = create_node
            create_node.next = self.head
            return
        while current.next is not self.head:
            current = current.next
        current.next = create_node
        create_node.next = self.head
        
        
    def traverse(self):
        current = self.head
        if not current:
            print('Circular Singly Linked List is empty')
            return
        
        while True:
            print('Data is :',current.data)
            current = current.next
            if current is self.head:
                break          
            
            
circular_linked_list = CircularSLinkedList()
circular_linked_list.traverse()
print("+++++++++++++++++++")
for i in range(5):
    circular_linked_list.append(i)
print("+++++++++++++++++++")
circular_linked_list.traverse()
```
