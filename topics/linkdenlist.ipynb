```python
# Finding Lowest value in linkden List
from math import inf
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def append(self,data):
        create_node = Node(data)
        if self.head is None:
            self.head = create_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = create_node 
    def find_lowest_value(self):
        current  = self.head
        if not current:
            return 'Empty Linkden List'
        lowest = float(inf)
        while current:
            cur_val = current.data
            if cur_val < lowest:
                lowest = cur_val
            current = current.next
        return lowest
        
            
linkdenlist = LinkedList()
for i in range(4):
    linkdenlist.append(i)
linkdenlist.find_lowest_value()

```

```python
# Delete a  Node by value in a  linked List
from math import inf
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def append(self,data):
        create_node = Node(data)
        if self.head is None:
            self.head = create_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = create_node 
    
    def tranverse(self):
        current = self.head
        if not current:
            print('Empty linked List')
            return
        while current:
            print(current.data)
            current = current.next
    
    def delete_a_node_by_value(self,val):
        current = self.head
        if not current:
            print('Empty Linked list')
            return
        if current.data == val:
            self.head = current.next
            return 
        while current.next:
            if current.next.data == val:
                current.next = current.next.next 
                return               
            current = current.next   
        
            
                     
linked_list = LinkedList() 
for i in [3,2,7]:
    linked_list.append(i)
    
print('+++++++++++++++++++++++++++++++++')
linked_list.tranverse()

print('+++++++++++++++++++++++++++++++++')
linked_list.delete_a_node_by_value(9)

print('+++++++++++++++++++++++++++++++++')
linked_list.tranverse()

```
