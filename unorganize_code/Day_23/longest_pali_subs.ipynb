```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummy = ListNode()  # Dummy head to simplify code
        current = dummy
        carry = 0

        # Loop while there is a node in l1 or l2 or a carry
        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0  # Value from l1 or 0
            val2 = l2.val if l2 else 0  # Value from l2 or 0

            total = val1 + val2 + carry
            carry = total // 10  # Update carry
            current.next = ListNode(total % 10)  # Store last digit
            current = current.next

            # Move to next nodes
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next  # Return the real head of the sum list
    
    
    
    
    
solution = Solution()
l1 = ListNode(2, ListNode(4, ListNode(3)))
l2 = ListNode(5, ListNode(6, ListNode(4)))
result = solution.addTwoNumbers(l1, l2)
print(result)  # Output: 7
```
