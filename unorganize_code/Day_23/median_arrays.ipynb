```python
from typing import List
```

```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        merged_array = nums1 + nums2
        merged_array.sort()
        n = len(merged_array)
        if n % 2 == 1:
            return float(merged_array[n // 2])
        else:
            mid1 = n // 2
            mid2 = mid1 - 1
            return (merged_array[mid1] + merged_array[mid2]) / 2.0
        
        
        
        
sol = Solution()
num1 = [1,2,3]
num2 = [2,3,4]
print(sol.findMedianSortedArrays(num1, num2))
        
```
